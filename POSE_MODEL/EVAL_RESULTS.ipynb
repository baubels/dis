```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
test_res = pd.read_csv('swept_tree77_test.csv')
min_bin = 0
max_bin = 180
nbins = 180
plt.figure(figsize=(16, 4))
plt.title('Histogram of Mean Angular Error on the test set of the last 80% of training')
data = test_res[int(0.98*len(test_res)):]['swept-tree-77 - Test_Mean_Angular']*180/np.pi
plt.hist(data, bins=nbins, range=(min_bin, max_bin), color='blue', alpha=0.5, label='Test');
plt.xlabel('Mean Angular Error (degrees)')
plt.ylabel('Number of images')
plt.legend(loc='upper right')

q1, q2, q3 = np.quantile(data, (0.25, 0.5, 0.75))
print('Q1: {:.2f} degrees'.format(q1))
print('Q2: {:.2f} degrees'.format(q2))
print('Q3: {:.2f} degrees'.format(q3))

plt.savefig('swept_tree77_test.png', dpi=300)
plt.show()
```

```python
test_res = pd.read_csv('swept_tree77_test_dis.csv')
min_bin = 0
max_bin = 180
nbins = 180
plt.figure(figsize=(16, 4))
plt.title('Histogram of Mean L2 distance prediction error on the test set of the last 20% of training')
data = test_res[int(0.8*len(test_res)):]['swept-tree-77 - Test_Mean_Distance']*180/np.pi
plt.hist(data, bins=nbins, range=(min_bin, max_bin), color='blue', alpha=0.5, label='Test');
plt.xlabel('Mean Angular Error (degrees)')
plt.ylabel('Number of images')
plt.legend(loc='upper right')

q1, q2, q3 = np.quantile(data, (0.25, 0.5, 0.75))
print(q1, q2, q3)

plt.savefig('swept_tree77_test.png', dpi=300)
plt.show()
```
